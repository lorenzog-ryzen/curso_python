- **IMC**: Se utilizará la columna de **IMC calculada** (resaltada en verde, basada en valores de peso y talla en la base de datos).
    - Se excluyen del análisis los registros con diferencias en IMC mayores de 2 o menores de -2:
      - ID: `20438765`, `17591769`, `41783045`, `28837359`, `3297269`, `79314863`.

```python
sub_temp = []
index_drop = []
for Id in [20438765, 17591769, 41783045, 28837359, 3297269, 79314863]:
    rows = df[df['individuo'] == Id]
    indices = rows.index.tolist()
    index_drop.append(indices)
    sub_temp.append(rows)
    df= df.drop(int(indices[0]))
    print(f"Indices para Id {Id}: {indices}")
result = pd.concat(sub_temp, ignore_index=False)
result
```

```python
df.reset_index(drop=True)
```
