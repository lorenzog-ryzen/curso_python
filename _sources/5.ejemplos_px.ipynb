##  1. Gráfico de Líneas

---

## 2. Gráfico de Dispersión (Scatter Plot)

```python
fig = px.scatter(df, x='edad', y='ingresos', color='genero', title='Edad vs Ingresos')
fig.show()
```

---

## 3. Gráfico de Barras

```python
fig = px.bar(df, x='ciudad', y='ventas', color='producto', title='Ventas por Ciudad')
fig.show()
```

---

## 4. Gráfico de Barras Apiladas

```python
fig = px.bar(df, x='ciudad', y='ventas', color='producto', title='Ventas por Ciudad', barmode='stack')
fig.show()
```

---

## 5. Gráfico de Pastel (Pie Chart)

```python
fig = px.pie(df, names='categoría', values='ventas', title='Participación por Categoría')
fig.show()
```

---

## 6. Box Plot (Diagrama de Caja)

```python
fig = px.box(df, x='genero', y='ingresos', title='Distribución de Ingresos por Género')
fig.show()
```

---

## 7. Histograma

```python
fig = px.histogram(df, x='edad', nbins=20, title='Distribución de Edad')
fig.show()
```

---

## 8. Mapa de Dispersión Geográfica

```python
fig = px.scatter_geo(df, lat='latitud', lon='longitud', color='pais', title='Ubicación de Puntos')
fig.show()
```

---

## 9. Gráfico de Área

```python
fig = px.area(df, x='fecha', y='ventas', color='producto', title='Ventas acumuladas por producto')
fig.show()
```

---

## 10. Gráfico de Línea con múltiples variables

```python
fig = px.line(df, x='fecha', y=['ventas_A', 'ventas_B'], title='Comparación de ventas A vs B')
fig.show()
```
