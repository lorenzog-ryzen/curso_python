# Estructuras de Control en Python

En este notebook aprenderemos sobre las estructuras de control más usadas en Python:

- **Ciclo `for`**
- **Ciclo `while`**
- **Condiciones `if`**

Estas estructuras nos permiten controlar el flujo de ejecución del programa.

## Ciclo `for`
El ciclo `for` se utiliza para iterar sobre una secuencia.

```python

frutas = ['manzana', 'banana', 'cereza']
for fruta in frutas:
    print('Fruta:', fruta)
```

```python
# Ejemplo de ciclo for con range
for i in range(5):
    print('Número:', i)
```

## Ciclo `while`
El ciclo `while` se ejecuta mientras una condición sea verdadera.

```python
# Ejemplo de ciclo while
contador = 0
while contador < 5:
    print('Contador:', contador)
    contador += 1
```

## Condición `if`
La estructura `if` permite ejecutar código basado en condiciones.

```python
# Ejemplo de if-elif-else
edad = 18
if edad < 18:
    print('Eres menor de edad')
elif edad == 18:
    print('Tienes 18 años')
else:
    print('Eres mayor de edad')
```

## Conclusión
En este notebook hemos aprendido sobre ciclos y estructuras condicionales en Python, con ejemplos de `for`, `while` y `if`.
