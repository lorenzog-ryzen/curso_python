## Ejemplos de codigos 

### Python
```
n = 10
for i in range(n):
    print(f'{n}**{i}: {n**i}')
```


### Java
```
public class Potencias {
    public static void main(String[] args) {
        int n = 10;
        for (int i = 0; i < n; i++) {
            System.out.println(n + "**" + i + ": " + (int) Math.pow(n, i));
        }
    }
}
```

### C++
```
#include <iostream>
#include <cmath>

int main() {
    int n = 10;
    for (int i = 0; i < n; i++) {
        std::cout << n << "**" << i << ": " << static_cast<int>(pow(n, i)) << std::endl;
    }
    return 0;
}
```

### JavaScript
```
let n = 10;
for (let i = 0; i < n; i++) {
    console.log(`${n}**${i}: ${n ** i}`);
}
```
